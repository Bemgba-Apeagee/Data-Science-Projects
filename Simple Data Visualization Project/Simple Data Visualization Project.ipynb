```python
import seaborn as sns
sns.get_dataset_
```

```python
iris = sns.load_dataset('iris')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
#sns.set(style='whitegrid')
iris.head()
```

```python
sns.regplot(x='sepal_length', y='sepal_width', data=iris)
```

```python
sns.lmplot(x='sepal_length', y='sepal_width', data=iris)
```

```python
sns.regplot(x='petal_length', y='petal_width', data=iris)
```

```python
sns.scatterplot(data=iris, x='sepal_length', y='sepal_width')
```

```python
sns.scatterplot(data=iris, x='sepal_length', y='sepal_width', hue='species')
```

```python
sns.scatterplot(data=iris, x='petal_length', y='petal_width')
plt.title('Relationship between petal length vs petal width')
```

```python
sns.scatterplot(data=iris, x='petal_length', y='petal_width', hue='species')
```

```python
# categorical scatter plot for variables
```

```python
sns.catplot(data=iris, x='species', y='sepal_length')
```

```python
sns.catplot(data=iris, x='species', y='sepal_length',jitter=False)
```

```python
sns.catplot(data=iris, x='species', y='sepal_length', kind='swarm')
```

```python
sns.catplot(data=iris, y='species', x='sepal_length',kind='swarm')
```

```python
# Box plot of variables
```

```python
sns.catplot(data=iris, x='species', y='sepal_length', kind='box')
plt.title('box plot for sepal length')
```

```python
sns.catplot(data=iris, x='species', y='sepal_width', kind='box')
plt.title('box plot for sepal width vs species')
```

```python
sns.catplot(data=iris, x='species', y='petal_length', kind='box')
plt.title('box plot for petal length vs species')
```

```python
sns.catplot(data=iris, x='species', y='petal_width', kind='box')
plt.title('box plot for petal width vs species')
```

```python
# Histogram plots
```

```python
sns.catplot(data=iris, x='species', y='sepal_length', kind='bar')
plt.title('bar plot of sepal length vs species')
```

```python
sns.catplot(data=iris, x='species', y='sepal_width', kind='bar')
plt.title('bar plot of sepal width vs species')
```

```python
sns.catplot(data=iris, x='species', y='petal_length', kind='bar', palette='winter')
plt.title('bar plot of petal length vs species')
```

```python
sns.catplot(data=iris, x='species', y='sepal_width', kind='bar', palette='summer')
plt.title('bar plot of sepal width vs species')
```

```python
sns.catplot(data=iris, x='species', kind='count', palette='ocean')
```

```python
# violin plots of the variables
```

```python
sns.catplot(data=iris, x = 'sepal_length', y='species', kind='violin')
plt.title('violin plot of sepal length vs species')
```

```python
sns.catplot(data=iris, y = 'sepal_width', x='species', kind='violin')
plt.title('violin plot of sepal width vs species')
```

```python
sns.catplot(data=iris, x = 'species', y='petal_length', kind='violin', palette='winter')
plt.title('violin plot of petal length vs species')
```

```python
sns.catplot(data=iris, x='species', y='petal_width', kind='violin', palette='summer')
plt.title('violin plot of petal width vs species')
```
