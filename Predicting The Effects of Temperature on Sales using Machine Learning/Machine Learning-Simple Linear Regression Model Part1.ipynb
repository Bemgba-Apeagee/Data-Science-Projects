#### Loading the Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

#### Imorting the dataset

```python
sales = pd.read_csv('temp_sales.csv')
sales.head()
```

```python
# data preprocessing checks
sales.isna().sum()
```

```python
sales.shape
```

#### Split the dataset into the Training Set and Test Set

```python
X = sales.iloc[:, :-1].values # indepedent variable
y = sales.iloc[:, -1].values # dependent variable
```

```python
X[0:10]
```

```python
y[0:10]
```

#### Training the Simple Linear Regression Model on the Training Set

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=1/3, random_state=0)
```

```python
X_train[0:10]
```

```python
X_test[0:10]
```

```python
y_train[0:10]
```

```python
y_test[0:10]
```

```python
# training the model

from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

#### Predicting the Test Set Results

```python
y_pred = regressor.predict(X_test)
y_pred
```

#### Visualizing the Training Set Results

```python
plt.scatter(X_train, y_train, color='red')
plt.plot(X_train, regressor.predict(X_train), color='blue')
plt.title('Effects of Temperature on Sales (Training Set)')
plt.xlabel('Temperature')
plt.ylabel('Sales')
plt.show()
```

#### Visualizing the Test Set Results

```python
plt.scatter(X_test, y_test, color='red')
plt.plot(X_test, regressor.predict(X_test), color='blue')
plt.title('Effects of Temperation on Sales (Test Set)')
plt.xlabel('Temperature')
plt.ylabel('Sales')
plt.show()
```

#### Decsision Tree for High Resolution

```python
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor(random_state=0)
regressor.fit(X, y)
```

```python
# Reshaping the dataset
X_grid = np.arange(max(X), min(X), 0.01)
x_grid = X_grid.reshape(len(X_grid), 1)
plt.scatter(X, y)
plt.plot(X, regressor.predict(X), color='red')
plt.title('Temperature vs Sales (Decision Tree)')
plt.xlabel('Temperature')
plt.ylabel('Sales')
plt.show()
```

#### Predict the result

```python
y_pred = regressor.predict([[98]])
y_pred
```

#### Evaluating the model

```python
from sklearn.metrics import mean_squared_error
```

```python
# Evaluating the model using mean saquared error
regressor.score(X_test, y_test)
```

- Our model has a high accuracy score of 0.98 (98%) in correctly predicting the effects of temperature on sales
