#### Welcome! This is a simple data visualization project that explores Penguins diverse characteristics derived from the Seaborn Penguins dataset.
The aim of the project is to create visualizations that can help uncover insights and patterns in the Seaborn Penguins dataset, providing valuable information for further analysis and decision-making.


#### This visualization project intends to provide insights into the following questions.
1. How do different penguin species compare in terms of body measurements?
2. Is there a relationship between different body measurements of penguins?
3. How do body measurements differ between male and female penguins?
4. How is the distribution of penguin species in terms of their island habitats?
5. Is there a correlation between body measurements and penguin species?
6. How does the body mass of penguins vary across different islands?


#### Importing the Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
%matplotlib inline
```

#### Loading the dataset

```python
penguins_df = pd.read_csv('seaborn_penguins_dataset.csv')
penguins_df.head()
```

```python
penguins_df.isna().sum()
```

```python
penguins_df.dropna()
```

#### 1. How do different penguin species compare in terms of body measurements?

```python
# box plot visual
sns.catplot(data=penguins_df, x= 'species', y='bill_length_mm', kind='box')
plt.title('Box plot showing different body measurements for species')
```

```python
# violin plot
sns.catplot(data=penguins_df, x='species', y='flipper_length_mm', kind='violin')
```

```python
sns.catplot(data=penguins_df, x='species', y='body_mass_g', kind='swarm')
```

These plots allows for comparing for easy comaprison of the measurements between species

```python
sns.catplot(data=penguins_df, x='species', y='bill_depth_mm')
```

#### 2. Is there a relationship between different body measurements of penguins?

```python
# pair plot to visualize various relationships

sns.pairplot(penguins_df)
plt.title('Pair plots showing relationship between different body measurements of penguins')
```

#### 3. How do body measurements differ between male and female penguins?

```python
# gender comparison
sns.catplot(data=penguins_df, x='sex', y='bill_length_mm', kind='box')
plt.title('Box plot showing bill legnth measurements between male and female penguins')
```

```python
sns.catplot(data=penguins_df, x='sex', y='flipper_length_mm', kind='violin')
plt.title('Box plot showing flipper length measurement between male and female penguins')
```

```python
sns.catplot(data=penguins_df, x='sex', y='bill_depth_mm', kind='swarm')
plt.title('Box plot showing bill depth measurements between male and female penguins')
```

```python
sns.catplot(data=penguins_df, x='sex', y='body_mass_g')
plt.title('Box plot showing Body mass measurements between male and female penguins')
```

#### 4. How is the distribution of penguin species in terms of their island habitats?

```python
# penguins species distribtion
sns.catplot(data=penguins_df, x='species', hue='island', kind='count')
plt.title('Bar plot showing the distribution of Penguin species based on their islands')
```

#### 5. Is there a correlation between body measurements and penguin species?

```python
# correlation
df = penguins_df[['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g', 'species']]
corr = df.corr('pearson')
```

```python
sns.heatmap(data=corr, cmap='viridis', annot=True, linewidth=.5, center=0)
plt.title('Heatmap showing Correlation between body measurements and penguin species')
```

#### 6. How does the body mass of penguins vary across different islands?

```python
sns.violinplot(data=penguins_df, x='island', y='bill_length_mm')
plt.title('Distribution of bill_length across different islands')
```

```python
# box plot
sns.boxplot(data=penguins_df, x='island', y='flipper_length_mm')
plt.title('Distribution of flipper length across different islands')
```

```python
#swarm
sns.swarmplot(data=penguins_df, x='island', y='bill_depth_mm')
plt.title('Distribution of bill depth across different islands')
```

```python
sns.stripplot(data=penguins_df, x='island', y='body_mass_g')
plt.title('Distribution of bill_length across different islands')
```
